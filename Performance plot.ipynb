```python
import numpy as np
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25]
y = [3017, 5641, 8403, 11446, 12000]
```

```python
plt.plot(x, y, color='k', alpha=0.5)
plt.scatter(x, y, s=100, marker="s", color='k')
plt.xticks(x, [str(_) for _ in x], fontsize=28)
plt.yticks([_ for _ in [4000, 6000, 8000, 10000, 12000]], [_ for _ in ['4k', '6k', '8k', '10k', '12k']], fontsize=28)
plt.xlabel("No. of data generators", fontsize=40)
plt.ylabel("No. of examples/sec", fontsize=40)
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('plot.pdf')
```
